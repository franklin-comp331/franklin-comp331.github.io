# Syllabus: COMP 331: AI-Assisted Software Engineering

This syllabus is referenced in the [Course Orientation](course_orientation.ipynb) and provides the full course structure, learning outcomes, topics, grading, and weekly schedule. For tool installation and setup, see the [Tool Setup Guide](tool_setup.ipynb).

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Format:** 12-week  
**Credits:** 4  
**Prerequisite:** COMP 311 – Advanced Java Programming or equivalent programming experience


## Course Description
This course explores how software engineers can use AI-assisted tools intentionally while building controlled, evaluable coding-agent workflows. Students will use GitHub Copilot, Java, GitHub, testing tools, evaluation criteria, and structured documentation to move from basic AI-assisted coding toward a complete Controlled Coding Agent final project. Emphasis is placed on structured task input, eval-driven development, human oversight, risk analysis, reproducibility, and critically evaluating AI-generated outputs.

## Learning Outcomes
By the end of the course, students will be able to:
- Set up and verify a Java development environment with GitHub Copilot and GitHub-based submission workflows. (Apply)
- Document AI-assisted development decisions through weekly interaction logs. (Analyze)
- Translate feature requests into structured task specifications for AI-assisted implementation. (Create)
- Define and run evaluations that measure correctness and detect regressions in AI-generated outputs. (Evaluate)
- Integrate structured input, AI-assisted generation, evaluation, and human approval into a controlled coding-agent workflow. (Create)
- Analyze reliability, security, maintainability, and production-readiness risks in AI-assisted systems. (Analyze / Evaluate)
- Deliver a complete Controlled Coding Agent final project with documentation, evaluation evidence, and a reproducible demo. (Create)

## Topics Covered
- Environment setup and AI-assisted software engineering foundations
- GitHub Copilot, GitHub Education, Git, Java IDEs, and repository-based submission workflows
- Weekly AI interaction logging and evidence-based reflection
- Structured AI workflows and task specifications
- Eval-driven development, measurable correctness criteria, and regression detection
- Integrated AI system implementation with structured input, generation, and evaluation
- Risk analysis, reliability, security, maintainability, and production readiness
- Controlled coding-agent workflows with human oversight checkpoints
- Final Project proposal, architecture, partial implementation, evaluation strategy, complete system, and demo

## Textbooks & Materials
- Selected academic journal and conference papers on AI in software engineering
- Coursera specialization: Generative AI for Software Engineering
- Official GitHub Copilot documentation and tutorials
- Practical guides on prompt engineering and AI tooling focused on Copilot
- Supplementary readings on evaluation, software testing, reproducibility, responsible AI, and software security
