# Syllabus: Special Topics in Computer Science – AI-Assisted Software Engineering

This syllabus is referenced in the [Course Orientation](course_orientation.ipynb) and provides the full course structure, learning outcomes, topics, grading, and weekly schedule. For tool installation and setup, see the [Tool Setup Guide](tool_setup.ipynb).

**Course Code:** COMP 480  
**Format:** 12-week  
**Credits:** 4  
**Prerequisite:** COMP 311 – Advanced Java Programming or equivalent programming experience


## Course Description
This course explores the integration of artificial intelligence-driven coding assistants into the software engineering lifecycle. Students will learn how to leverage Copilot for tasks such as code generation, testing, debugging, documentation, and project management. Emphasis is placed on prompt engineering, human-AI collaboration, ethical considerations, security implications, and critically evaluating AI-generated outputs. Students will engage in practical labs and complete a culminating project that applies AI-assisted development in realistic software engineering scenarios.

## Learning Outcomes
By the end of the course, students will be able to:
- Recall and summarize key AI technologies and their roles within the software engineering lifecycle. (Remember / Understand)
- Compose effective prompts to generate, test, and debug Java code using AI-driven coding assistants. (Create)
- Apply AI-assisted tools to automate and enhance Java software development, testing, documentation, and project management tasks. (Apply)
- Evaluate the quality, reliability, security, and ethical implications of AI-generated software artifacts using defined criteria. (Evaluate)
- Collaborate with team members to design, implement, and deliver a comprehensive Java software project incorporating AI-assisted development practices. (Apply / Create)
- Analyze and critique the limitations, risks, and societal impacts of AI adoption in software engineering practice. (Analyze / Evaluate)

## Topics Covered
- Foundations of Software Engineering and AI-Driven Coding Assistants
- Overview of AI-Powered Coding Assistants and Their Integration in Java Development Workflows
- Prompt Engineering: Techniques and Best Practices for Java Code Generation
- Human-AI Collaboration in Software Development
- AI-Driven Java Code Generation and Automated Unit Test Creation (Using JUnit)
- Debugging and Error Resolution Assisted by AI Tools
- Automating Java Documentation and Code Review with AI Assistance
- Security Considerations and Risks of AI-Generated Java Code
- Project Management and Agile Software Development Practices Supported by AI Tools
- Ethical, Legal, and Societal Implications of AI-Assisted Software Engineering
- Future Trends and Emerging Technologies in AI-Assisted Software Engineering
