## 5) Collaboration, AI Use, and Academic Integrity


**Responsible AI Use**

- AI coding assistants (e.g., Copilot) are **allowed** for programming tasks **unless explicitly disallowed** for a specific assignment.

- Always **review, refactor, and test** AI-generated code—**you are responsible** for correctness and security.

- **Document AI assistance** in each submission (e.g., in README):

  - _"Used Copilot to draft `FooService.java` and `FooServiceTest.java`; I modified the error handling and added boundary tests."_



**Prohibited**

- Submitting unacknowledged AI-generated prose (e.g., essays) as your own

- Using AI to access/replicate proprietary or exam solutions

- Copy-pasting code from unknown sources without attribution/review



**Collaboration**

- Discuss concepts freely; **do not share full solutions** unless an assignment is explicitly team-based

- For team tasks, use issues/PRs to record contributions; follow code review norms

## 6) Ethics, Security, and Privacy (Primer)


**Key Risks to Watch**

- **Security:** Injections, unsafe deserialization, weak crypto, unvalidated inputs, secret leakage

- **Bias & Fairness:** Stereotypes encoded in training data; skewed recommendations

- **IP & Licensing:** Risk of copying licensed code snippets; verify provenance and licenses

- **Privacy:** Avoid pasting sensitive data into prompts; scrub secrets and PII



**Starter Reflection (Week 7):**

> In ~150–250 words, describe one risk you foresee when using AI to generate code and how you would mitigate it during development and code review.

## 7) Learning Activities & Participation


**Lab Pattern (most weeks)**

1. **Prompt**: Write a clear task prompt (include constraints & examples)

2. **Generate**: Use Copilot to propose code/tests/docs

3. **Critique**: Evaluate for correctness, complexity, style, security

4. **Iterate**: Improve prompt or code; retest; commit w/ notes



**Participation Signals**

- Attending/engaging in discussions or forums

- Offering constructive code review feedback

- Sharing prompt strategies and lessons learned

- Respectful teamwork and communication



---

## 8) Success Strategies


- Keep a **Prompt Journal**: record prompts, outputs, fixes, and what worked

- Prefer **small, precise prompts** over broad requests; iterate quickly

- Write **tests first** where feasible; use AI to scaffold edge cases

- Use **static analysis** and **security linters** alongside AI suggestions

- Commit early/often; narrate changes in PR descriptions

- Treat AI like a **junior pair-programmer**—great for drafts, not the final authority
