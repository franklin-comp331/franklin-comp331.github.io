## 5) AI Use, Evidence, and Academic Integrity

GitHub Copilot is the approved AI assistant for course programming work unless an assignment says otherwise. You may use Copilot to help draft code, tests, documentation, evaluation criteria, and workflow artifacts, but you remain responsible for every submitted artifact.

**Required Evidence**

- Record meaningful Copilot interactions in the [Weekly AI Interaction Log](../assignments/weekly_ai_interaction_log.ipynb).
- Describe AI assistance in each repository README or assignment documentation.
- Preserve prompts, accepted or rejected suggestions, manual edits, evaluations, and final decisions when relevant.

**Not Allowed**

- Submitting AI-generated work without acknowledgment.
- Submitting code you do not understand or cannot explain.
- Sharing full solutions with classmates unless the professor explicitly makes an activity collaborative.
- Pasting secrets, private data, proprietary code, or protected information into AI tools.

## 6) Risks To Watch Throughout the Course

The labs and Final Project ask you to evaluate AI-assisted systems, not just produce code. Keep these risks visible in your weekly logs, evaluations, and project documentation.

- **Correctness:** Generated code may appear plausible while missing edge cases or requirements.
- **Security:** AI suggestions may introduce unsafe input handling, secret leakage, weak cryptography, or injection risks.
- **Maintainability:** Generated code may be hard to read, overcomplicated, or poorly aligned with your architecture.
- **Evaluation gaps:** Passing one demo does not prove the workflow is reliable. Define measurable checks.
- **Human oversight:** Controlled workflows need clear points where a human reviews, approves, rejects, or revises AI output.

## 7) Learning Activities


**Controlled Workflow Pattern**

1. **Specify**: Translate a request into structured input with goals, constraints, edge cases, and acceptance criteria

2. **Generate or revise**: Use Copilot to propose implementation, tests, documentation, or workflow artifacts

3. **Evaluate**: Run tests, scripts, checks, or structured reviews against explicit criteria

4. **Control**: Add human approval checkpoints before accepting important changes

5. **Document**: Preserve prompts, outputs, evaluation results, risks, and final decisions


This pattern appears repeatedly in Labs 1-5 and the Final Project. The goal is not to maximize AI usage; the goal is to make AI-assisted development controlled, observable, evaluable, and reproducible.


---

## 8) Success Strategies


- Keep the **Weekly AI Interaction Log** current: record prompts, outputs, evaluations, decisions, and what worked

- Prefer **small, precise prompts** over broad requests; iterate quickly

- Define evaluation criteria before trusting generated code; use AI to scaffold edge cases where appropriate

- Use tests, static analysis, security checks, and manual review alongside AI suggestions

- Commit early/often; preserve artifacts that make workflow decisions reproducible

- Treat AI like a **junior pair-programmer**: useful for drafts, but never the final authority
