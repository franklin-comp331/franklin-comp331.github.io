# Tool Setup Guide: Java & GitHub Copilot

This guide is referenced in the [Course Orientation](course_orientation.ipynb) and is required for Lab 1. For full course structure and grading, see the [Syllabus](syllabus.ipynb).

Choose your preferred IDE and follow the full setup walkthrough below.

## Visual Studio Code Setup

1. **Install VS Code**
   - Download from [https://code.visualstudio.com/](https://code.visualstudio.com/).
   - Install and launch VS Code.

2. **Install Java & Extensions**
   - Download and install the latest JDK from [Adoptium](https://adoptium.net/) or [Oracle](https://www.oracle.com/java/technologies/downloads/).
   - Open Extensions view (`Cmd+Shift+X`/`Ctrl+Shift+X`).
   - Install: Extension Pack for Java, Java Test Runner, Debugger for Java, Maven for Java, GitHub Copilot.

3. **Configure Java**
   - Open Command Palette and select `Java: Configure Java Runtime`.
   - Ensure the correct JDK is selected.

4. **Create a Java Project**
   - Open Command Palette and select `Java: Create Java Project`.
   - Choose a build tool (No build tools, Maven, or Gradle).
   - Select a location and open the generated project.

5. **Test Java Code Execution**
   - Create a file `HelloWorld.java` and add:
     ```java
     public class HelloWorld {
         public static void main(String[] args) {
             System.out.println("Hello, world! Your Java environment is working.");
         }
     }
     ```
   - Compile: `javac HelloWorld.java`
   - Run: `java HelloWorld`

6. **Run Unit Tests**
   - Create `CalculatorTest.java` and add:
     ```java
     import static org.junit.Assert.*;
     import org.junit.Test;
     public class CalculatorTest {
         @Test
         public void testAddition() {
             assertEquals(2, 1 + 1);
         }
     }
     ```
   - Run tests using Java Test Runner or `mvn test`/`gradle test`.

7. **Get Started with GitHub Copilot**
   - Install Copilot extension.
   - Sign in with GitHub.
   - Start typing code for AI suggestions.
   - Accept suggestions with `Tab`.


## IntelliJ IDEA Setup

1. **Install IntelliJ IDEA**
   - Download from [https://www.jetbrains.com/idea/download/](https://www.jetbrains.com/idea/download/).
   - Choose Professional Edition (free for students with school email).
   - Install and launch IntelliJ IDEA.

2. **Install Java & Plugins**
   - Download and install the latest JDK from [Adoptium](https://adoptium.net/) or [Oracle](https://www.oracle.com/java/technologies/downloads/).
   - Go to Preferences > Plugins.
   - Choose the Marketplace tab to search for plugins.
   - Install: GitHub Copilot, JUnit (usually included), Maven or Gradle (usually included).

3. **Configure Java**
   - Go to Preferences > Build, Execution, Deployment > Build Tools > Gradle/Maven.
   - Ensure the correct JDK is selected.

4. **Create a Java Project**
   - On the Welcome screen, click "New Project".
   - Select Java, choose Maven or Gradle if desired.
   - Set project location and settings.
   - Create a file `HelloWorld.java` and add:
     ```java
     public class HelloWorld {
         public static void main(String[] args) {
             System.out.println("Hello, world! Your Java environment is working.");
         }
     }
     ```
   - Run the file using the green play button or right-click > Run.

5. **Run Unit Tests**
   - Create `CalculatorTest.java` and add:
     ```java
     import static org.junit.Assert.*;
     import org.junit.Test;
     public class CalculatorTest {
         @Test
         public void testAddition() {
             assertEquals(2, 1 + 1);
         }
     }
     ```
   - Run tests using the built-in test runner.

6. **Get Started with GitHub Copilot**
   - Install Copilot plugin.
   - Sign in with GitHub.
   - Start typing code for AI suggestions.
   - Accept suggestions with `Tab`.
