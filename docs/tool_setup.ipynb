# Tool Setup Guide: Java, Git, and GitHub Copilot

This guide helps you get the required tools installed and opened for Week 1. After completing these checks, use [Assignment 1](../assignments/assignment01.ipynb) to practice with the tools, create your repository, capture screenshots, and complete the reflection.

The tool order below matches [Assignment 1](../assignments/assignment01.ipynb): JDK, IDE, Git/GitHub, Copilot, then a simple Java project.

## 1. Install JDK 17+

Install a JDK from one of these sources:

- [Adoptium Temurin](https://adoptium.net/)
- [Oracle Java Downloads](https://www.oracle.com/java/technologies/downloads/)

Confirm Java is available from a terminal:

```bash
java -version
```

You should see version 17 or newer.

## 2. Install One IDE

Choose one development environment for the course. You do not need both.

**Option A: IntelliJ IDEA**

- Download IntelliJ IDEA from [JetBrains](https://www.jetbrains.com/idea/download/).
- JetBrains recently merged the Community Edition into the main IntelliJ IDEA product offering. Students can also check the [JetBrains Student Pack](https://www.jetbrains.com/academy/student-pack/) for access options.
- Open IntelliJ and confirm it can create or open a Java project.

**Option B: Visual Studio Code**

- Download VS Code from [code.visualstudio.com](https://code.visualstudio.com/).
- Open Extensions (`Cmd+Shift+X` on macOS or `Ctrl+Shift+X` on Windows/Linux).
- Install **Extension Pack for Java**.

Maven, Gradle, and JUnit are useful later in the course, but they are not required for the basic Assignment 1 setup check.

## 3. Set Up Git and GitHub

Assignment 1 asks you to submit a Git repository link, so make sure Git and GitHub are ready before you start the assignment work.

1. Install Git from [git-scm.com](https://git-scm.com/downloads) if it is not already installed.
2. Confirm Git is available:

```bash
git --version
```

3. Sign in to [GitHub](https://github.com/) or create an account.
4. Set up GitHub Education using your university email address at [github.com/settings/education/benefits](https://github.com/settings/education/benefits). This is how students access GitHub Education benefits, including Copilot access when approved.
5. Confirm you know how you will connect to GitHub: HTTPS or SSH. GitHub's docs explain both options in [Set up Git](https://docs.github.com/en/get-started/git-basics/set-up-git).

## 4. Enable GitHub Copilot

GitHub Copilot is the only AI assistant approved for Assignment 1. Before installing the IDE plugin, request or confirm GitHub Education benefits with your university email address at [github.com/settings/education/benefits](https://github.com/settings/education/benefits).

**In IntelliJ IDEA**

- Open Settings/Preferences > Plugins.
- Search for **GitHub Copilot** in the Marketplace and install it.
- Restart IntelliJ if prompted.
- Sign in with your GitHub account when prompted.

**In VS Code**

- Open Extensions.
- Install **GitHub Copilot**.
- Sign in with your GitHub account when prompted.

If you need access help, see [GitHub Copilot documentation](https://docs.github.com/en/copilot).

## 5. Create a Simple Java Project

Create or open a small Java project in your IDE. For Assignment 1, name the project:

```text
ai-swe-course
```

Add a basic `Main.java` file and confirm it runs. A minimal starting point is:

```java
public class Main {
    public static void main(String[] args) {
        System.out.println("Java setup is working.");
    }
}
```

Use your IDE's Run button or run from the terminal if your file is in the current directory:

```bash
javac Main.java
java Main
```

## 6. Confirm Copilot Suggestions Appear

In `Main.java`, type a short comment such as:

```java
// method to add two numbers
```

Pause briefly and look for Copilot ghost text. If a suggestion appears, accept it with `Tab` or your IDE's configured accept-suggestion shortcut.

Do not spend time making this code complex. The goal is only to confirm that Java, your IDE, GitHub Copilot, and GitHub are ready.

## 7. Continue in Assignment 1

This setup guide stops once the tools are working. Complete the actual Week 1 deliverables in [Assignment 1: Environment Setup & AI-Assisted SWE Foundations](../assignments/assignment01.ipynb).

Assignment 1 is where you will:
