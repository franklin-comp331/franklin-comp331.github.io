**Course Name/Number:** COMP 331: AI-Assisted Software Engineering

## Purpose of Assignment

This assignment is intended to help you:

- Implement the core agent loop
- Design controlled multi-step workflows
- Introduce human oversight into your coding agent
- Move from simple workflows to agent-like behavior


## Steps / Requirements of Assignment

Implement a controlled multi-step workflow that resembles an agent loop while keeping human oversight explicit. The system should not blindly generate and apply changes without review.

1. Define the loop stages, such as intake, plan, generate, evaluate, revise, and approve.
2. Implement the workflow in code, scripts, or a clearly runnable process.
3. Add at least one human approval checkpoint.
4. Preserve artifacts from a workflow run, including input, generated output, evaluation result, and final decision.
5. Document how to reproduce the workflow.


## Action Items

- Implement the workflow in your repository.
- Document how the workflow operates.
- Ensure reproducibility.
- Record relevant AI interactions in your final Week 10 AI interaction log entry.


## Deliverables

Your repository should include:

- Runnable controlled workflow.
- Documentation of the agent loop stages.
- Human oversight checkpoint description.
- Example run artifacts.
- Reproducible setup and execution instructions.
