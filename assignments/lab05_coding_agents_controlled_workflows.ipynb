## Steps / Requirements of Assignment

Implement a controlled multi-step workflow that resembles an agent loop while keeping human oversight explicit. The system should not blindly generate and apply changes without review. By the end of this lab, your Java project should behave like a small, safe version of a coding agent: it receives a task, lets the LLM request approved tools, feeds tool results back into the conversation, and stops with a final answer or a safety limit.

1. Define the loop stages in your README before coding. Recommended stages: intake, model request, tool-call detection, approval, tool execution, tool-result message, evaluation, final response.
2. Define tool schemas or tool descriptions for the Java tools you built in Lab 3. Include at least file listing, file reading, file writing, and Java test execution.
3. Register available tools in a dispatcher, map, or registry so that a requested tool name can be matched to the correct Java implementation.
4. Send the tool declarations or equivalent tool descriptions with the LLM request using your chosen API. If your API wrapper does not support native function calling, you may use a structured JSON convention, but you must document the format clearly.
5. Detect when the model requests a tool call. Extract the tool name and arguments.
6. Before executing any write or command-running tool, add a human approval checkpoint. This can be an interactive `y/n` prompt, a `--auto-approve-readonly` mode, or a documented manual review step. File reads and file listings may be allowed automatically if your risk analysis supports that choice.
7. Execute allowed tool calls using your Lab 3 tool implementations. Unknown tools, invalid arguments, unsafe paths, and blocked commands must produce safe error messages rather than crashes.
8. Append tool results back into the message history and send the updated conversation back to the model.
9. Repeat the loop until the model produces a final answer or until a maximum iteration limit is reached. Recommended maximum: 10-20 iterations.
10. Preserve artifacts from one complete workflow run. Include the user prompt, model responses, tool calls, tool outputs, approval decisions, changed files, evaluation result, and final response.
11. Demonstrate a complete controlled workflow using your sandbox project. A good example is: inspect the calculator project, identify a bug or missing behavior, propose a change, request approval, write the change, run tests, and report the result.
12. Document how to reproduce the workflow from a clean checkout.

Suggested pseudocode:

```java
messages.add(Message.user(userPrompt));
for (int i = 0; i < maxIterations; i++) {
    ModelResponse response = client.generate(messages, toolSchemas);
    messages.add(response.asMessage());

    if (!response.hasToolCall()) {
        System.out.println(response.finalText());
        break;
    }

    ToolCall call = response.toolCall();
    if (requiresApproval(call) && !approvedByHuman(call)) {
        messages.add(Message.toolResult(call.name(), "Tool call rejected by human reviewer."));
        continue;
    }

    String result = dispatcher.execute(call);
    messages.add(Message.toolResult(call.name(), result));
}
```
