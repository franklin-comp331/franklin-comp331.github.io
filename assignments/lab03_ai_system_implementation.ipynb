**Course Name/Number:** COMP 331: AI-Assisted Software Engineering

## Purpose of Assignment

This assignment is intended to help you:

- Integrate components into a working coding agent workflow
- Connect structured input, AI generation, and evaluation
- Transition from isolated tasks to a cohesive system


## Steps / Requirements of Assignment

Integrate your structured input and evaluation work into a cohesive workflow. The system does not need to be fully autonomous, but it should show a repeatable path from task request to generated or revised code to evaluation.

1. Define the main workflow stages.
2. Connect the task specification to code generation or revision.
3. Connect the generated output to your evaluation layer.
4. Add clear instructions or a command that runs the workflow.
5. Document the architecture and current limitations.


## Action Items

- Build the system in your repository.
- Ensure the system is runnable.
- Document the architecture.
- Record prompts, iterations, and evaluation outcomes in your weekly log.


## Deliverables

Your repository should include:

- Working source code for the integrated workflow.
- Architecture documentation in the README or a separate design document.
- A sample task input and resulting output.
- Evaluation results from at least one workflow run.
