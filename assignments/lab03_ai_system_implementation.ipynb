- Integrate components into a working coding agent workflow

- Transition from isolated tasks to a cohesive system
- Implement safe Java tools that allow an agent to inspect and work inside a sandbox


## Steps / Requirements of Assignment

Integrate your structured input and evaluation work by adding Java tool implementations. These tools are the controlled operations your coding agent will later expose to the LLM. At this stage, you may test the tools directly from Java code or JUnit; the model does not need to call them automatically yet.

1. Define a common tool interface or base pattern. For example, you might create an `AgentTool` interface with methods such as `name()`, `description()`, and `execute(...)`.
2. Define one sandbox working directory. This should point to the Java sandbox project from Lab 2, such as `workspace/calculator`.
3. Implement a file listing tool similar to `get_files_info`. It should list files and directories plus basic metadata such as size and whether each entry is a directory.
4. Implement a file reading tool similar to `get_file_content`. It should read text files from the sandbox and cap output at a reasonable length, such as 10,000 characters.
5. Implement a file writing tool similar to `write_file`. It should create or overwrite files only inside the sandbox.
6. Implement a command execution tool similar to `run_python_file`, adapted for Java. Recommended: `RunJavaTestsTool` that runs `mvn test` or `gradle test` in the sandbox. Avoid arbitrary shell access unless you have a strict allowlist.
7. Use Java `Path` APIs to prevent path traversal. Paths such as `../README.md` or absolute paths outside the sandbox must be rejected.
8. Add JUnit tests for each tool. Include successful cases and blocked cases for missing files, invalid directories, path traversal, and command restrictions.
9. Add a small manual driver or test command that demonstrates the tools working together: list files, read a Java file, write a harmless file, and run the sandbox tests.
10. Document the architecture and current limitations in your README.
