# Lab 1: Structured AI Workflow

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Week:** 3  
**Points:** 90

Continue maintaining the [Weekly AI Interaction Log](weekly_ai_interaction_log.ipynb) while completing this lab.


## Purpose of Assignment

This assignment is intended to help you:

- Establish the task input layer of your coding agent
- Translate feature requests into structured specifications
- Compare naive vs structured AI-assisted development
- Begin building the foundation of your agent system


## Steps / Requirements of Assignment

Build the first layer of a Controlled Coding Agent workflow: a structured task input process. Use GitHub Copilot intentionally, but keep human judgment visible in your documentation.

1. Choose a small Java feature request that can be implemented in a short program or utility class.
2. Implement the feature once using a naive prompt or informal comment.
3. Create a structured task specification that includes goal, inputs, outputs, constraints, edge cases, and acceptance criteria.
4. Re-implement or revise the feature using the structured specification.
5. Compare the two approaches in your README.


## Action Items

- Create a Git repository or use your continuing course repository.
- Implement both the naive and structured approaches.
- Include documentation in `README.md`.
- Record relevant Copilot interactions in your weekly AI interaction log.


## Deliverables

Your repository should include:

- Java source code for both approaches or clearly separated commits showing each approach.
- A structured task specification.
- A `README.md` explaining how to run the code and comparing naive vs structured development.
- Evidence of AI-assisted development through prompts, comments, screenshots, or log links.
