## Purpose of Assignment

This assignment is intended to help you:

- Establish the task input layer of your coding agent
- Translate feature requests into structured specifications
- Compare naive vs structured AI-assisted development
- Begin building the foundation of a Java-based coding agent system


## Steps / Requirements of Assignment

Build the first layer of a Controlled Coding Agent workflow: a Java command-line program that can accept a user request, send it to an LLM, and compare naive prompting with structured prompting. This lab should feel small and controlled. You are not building the full agent yet.

Start fresh for this lab. Create a new repository named `comp331-java-agent`. Do not use your Assignment 1 setup repository as the starting point unless your professor explicitly approves it. This Lab 1 repository should become your continuing repository for Labs 1-5 and may become the starting point for your final project.

1. Create a fresh GitHub repository named `comp331-java-agent`.
2. Create a Java project using Maven in that repository. A simple structure such as `src/main/java/.../Main.java` is enough.
3. Configure your Gemini API key using the [Gemini API Setup Guide](../docs/gemini_api_setup.ipynb). Do not commit API keys, `.env` files, or other secrets.
4. Write a `Main` class that first sends one hard-coded prompt to the model and prints the response. This verifies that your Java project can connect to the API.
5. Revise the program so it accepts the user prompt from command-line arguments, for example `mvn exec:java -Dexec.args="Explain this Java method"`.
6. Add a simple message representation for later labs. This may be a small `Message` class or record with fields such as `role` and `content`.
7. Add a `--verbose` flag or similar option that prints helpful debugging information, such as the user prompt, selected model, and token counts if your API provides them.
8. Choose a small Java feature request that can be implemented in a utility class, such as formatting a string, validating input, parsing a simple expression, or computing a value.
9. Ask GitHub Copilot or another approved coding assistant for help once using a naive prompt or informal comment. Save the prompt and resulting code or describe the interaction.
10. Create a structured task specification with goal, inputs, outputs, constraints, edge cases, and acceptance criteria. Use that specification to revise or re-implement the feature.
11. Compare the naive and structured approaches in your README. Focus on concrete differences in output quality, missing edge cases, and human decisions.

Suggested checkpoint commands:
