## Purpose of Assignment

This assignment is intended to help you:

- Build the evaluation layer of your coding agent
- Define measurable correctness criteria
- Detect regressions in AI-generated outputs
- Create the Java sandbox project that later agent tools will inspect and modify


## Steps / Requirements of Assignment

Extend your Lab 1 repository by adding a small Java sandbox project and a repeatable evaluation process. Later labs will give the coding agent tools that operate on this sandbox, so keep it simple and easy to test.

1. Create a sandbox Java project or package inside your repository. A recommended name is `workspace/calculator` or `sandbox/calculator`.
2. Implement a small Java program or library that has clear behavior. Recommended options include a calculator, string formatter, unit converter, password validator, or CSV summary utility.
3. Add JUnit tests that check normal cases, edge cases, and at least one invalid-input case.
4. Define measurable acceptance criteria in your README. Each criterion should connect to a test, command, or review check.
5. Run the evaluation manually from the command line using `mvn test` or an equivalent command. Save the output or summarize it in your README.
6. Use Copilot or another approved AI tool to make at least two iterations of the sandbox code. One iteration may intentionally be incomplete or flawed.
7. Run the evaluation against at least two versions or iterations of the code and document what changed. You may use separate commits, branches, screenshots, copied terminal output, or a written comparison.
8. Connect the evaluation back to your Lab 1 structured prompt/specification. Note which acceptance criteria were enforced automatically and which still required human review.

Suggested checkpoint commands:

```bash
mvn test
mvn -q test
git log --oneline
```

Minimum viable version: one sandbox Java class and at least five meaningful JUnit tests. Stretch goal: add a small script or Maven profile that runs only the sandbox evaluation.

## Using GitHub Copilot

Use GitHub Copilot to strengthen the evaluation process, not to bypass it. You remain responsible for deciding which tests matter and whether the implementation satisfies the assignment.

Recommended Copilot workflow:

- Ask Copilot Chat to propose JUnit tests before implementation. Review the suggestions and keep only meaningful cases.
- Ask Copilot to identify missing edge cases after your first test set is written.
- Use Copilot to help create one incomplete or flawed implementation, then use your tests to expose the weakness.
- Label tests or test ideas as human-authored, Copilot-assisted, or Copilot-suggested but modified.
- Record how each accepted Copilot suggestion was verified with `mvn test` or an equivalent command.
