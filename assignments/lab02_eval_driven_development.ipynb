# Lab 2: Eval-Driven Development

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Week:** 5  
**Points:** 100

This lab extends [Lab 1: Structured AI Workflow](lab01_structured_ai_workflow.ipynb). Continue maintaining the [Weekly AI Interaction Log](weekly_ai_interaction_log.ipynb).


## Purpose of Assignment

This assignment is intended to help you:

- Build the evaluation layer of your coding agent
- Define measurable correctness criteria
- Detect regressions in AI-generated outputs


## Steps / Requirements of Assignment

Extend your Lab 1 repository by adding an evaluation system for generated or revised code. The evaluation layer should make correctness criteria explicit and repeatable.

1. Identify the feature or task your evaluation will check.
2. Define measurable acceptance criteria.
3. Add automated checks, tests, scripts, or structured review criteria.
4. Run the evaluation against at least two versions or iterations of the code.
5. Document what the evaluation caught and what it did not catch.


## Action Items

- Extend your existing repository.
- Add an evaluation system.
- Document how to run evaluations.
- Record Copilot interactions and evaluation decisions in your weekly log.


## Deliverables

Your repository should include:

- Evaluation criteria in a documented format.
- Runnable evaluation code, tests, scripts, or checklist.
- Example evaluation results.
- README instructions for running the evaluations from a clean environment.
