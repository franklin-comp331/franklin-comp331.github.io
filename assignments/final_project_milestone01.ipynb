## Purpose of Assignment

This assignment is intended to help you:

- Define the scope of your Controlled Coding Agent
- Design a clear system architecture based on Labs 1-5
- Establish a structured AI-assisted workflow
- Create a realistic development plan for completion


## Steps / Requirements of Assignment

Prepare a proposal and architecture plan for your final Controlled Coding Agent system. You may continue your lab repository or create a dedicated final project repository.

1. Define the problem your coding agent addresses.
2. Describe intended users and expected workflow.
3. Create an architecture diagram or structured architecture description.
4. Identify core components, including input, generation, evaluation, workflow, and persistence.
5. Create a development plan for Weeks 10-12.
