# Final Project Milestone 1: Proposal & Architecture

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Week:** 10  
**Points:** 80

This milestone builds on Labs 1-5, especially [Lab 5: Coding Agents and Controlled Workflows](lab05_coding_agents_controlled_workflows.ipynb).


## Purpose of Assignment

This assignment is intended to help you:

- Define the scope of your Controlled Coding Agent
- Design a clear system architecture based on Labs 1-5
- Establish a structured AI-assisted team workflow
- Create a realistic development plan for collaborative completion


## Steps / Requirements of Assignment

Prepare a proposal, collaboration plan, and architecture plan for your final Controlled Coding Agent system. You may continue your lab repository or create a dedicated final project repository.

1. Define the problem your coding agent addresses.
2. Describe intended users and expected workflow.
3. Identify team members, responsibilities, asynchronous collaboration practices, communication expectations, and a task distribution plan for Weeks 10-12.
4. Evaluate at least two architecture alternatives, discuss tradeoffs, and choose one.
5. Create an architecture diagram or structured architecture description that maps responsibilities to components, services, or modules.
6. Identify core components, including input, generation, evaluation, workflow, and persistence.
7. Create a development plan for Weeks 10-12.


## Using GitHub Copilot

Use GitHub Copilot as the supported planning and critique partner, not as a substitute for your architecture decisions. Treat AI-generated code and documentation as untrusted until reviewed and validated. You remain responsible for scope, feasibility, and tradeoffs.

Recommended Copilot workflow:

- Ask Copilot Chat to critique your project scope for feasibility in the remaining course timeline.
- Ask Copilot to generate two or three architecture alternatives, then choose and justify one.
- Ask Copilot for likely risks, evaluation strategy ideas, and a milestone breakdown.
- Briefly document where Copilot assisted planning, which decisions required human judgment, and how conflicting suggestions were evaluated.
- Document at least one Copilot-generated alternative that you did not choose and explain why.
