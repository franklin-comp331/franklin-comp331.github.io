## Purpose of Assignment

This assignment is intended to help you:

- Demonstrate meaningful progress on your coding agent system
- Define and implement an evaluation strategy
- Identify risks and limitations in your system
- Prepare for final system completion


## Steps / Requirements of Assignment

Submit a partial implementation that demonstrates the core direction of your final project. The system does not need to be complete, but it must show meaningful progress and a concrete evaluation strategy.

1. Implement the most important workflow path or system component.
2. Add or refine the evaluation strategy.
3. Run the current system on at least one example task.
4. Identify current risks, limitations, and incomplete pieces.
5. Update your plan for the final submission.


## Using GitHub Copilot

Use GitHub Copilot for focused implementation, debugging, and evaluation work. Do not ask Copilot to generate the whole project at once; use it on scoped components and verify each result.

Recommended Copilot workflow:

- Use Copilot for one focused implementation task tied to your architecture.
- Use Copilot to improve or expand evaluation cases.
- Use Copilot to debug one failing example run, then document the fix and verification.
- Identify which implementation pieces were Copilot-assisted and how each was verified.
