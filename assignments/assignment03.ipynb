## 1. Overview
Apply AI-assisted debugging, documentation, and code-review techniques using GitHub Copilot. You will identify and fix bugs, generate and refine JavaDoc, and compare AI-assisted review feedback to human review practices.

**Outcomes:** improved debugging workflows, better documentation skills, and evidence-based critique of Copilot’s review suggestions.

---

## 2. Alignment with the Syllabus
- Use Copilot to identify and resolve bugs.
- Produce high-quality JavaDoc and inline documentation.
- Conduct AI-assisted code reviews and evaluate their usefulness.

## 3. Tooling & Policy

## 4. Tasks (What to do + Record)

### 1. Debugging with Copilot (40 pts)
What to do:
- Choose buggy Java code (own or provided) and use Copilot to suggest fixes and explanations.
- Produce before/after code and JUnit tests demonstrating correctness.
Record:
- Exact prompts, Copilot suggestions, final fix, and verification evidence (test output screenshots).

### 2. Auto-Documentation (30 pts)
What to do:
- Use Copilot to generate JavaDoc and improve inline comments for one class.
- Edit generated docs for accuracy and tone.
Record:
- Original vs. edited JavaDoc, reasons for edits, and a short assessment of quality.

### 3. AI-Assisted Code Review (40 pts)
What to do:
- Use Copilot to produce a review summary for a Java file.
- Compare its recommendations to a manual review and select actionable items.
Record:
- Copilot’s review output, three actionable items, and your agreement/disagreement rationale.

### 4. Reflection & Analysis (25 pts)
What to do:
- Write a ~300–400 word reflection on Copilot for debugging, documentation, and review.
Record:
- Final reflection and one concrete recommendation to improve Copilot’s usefulness.
