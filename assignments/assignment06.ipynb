## 1. Overview
Design, implement, test, and present a comprehensive Java project that integrates AI-assisted development practices. This capstone demonstrates synthesis of prompt engineering, secure/maintainable AI-assisted coding, testing, and professional documentation.

This year the capstone supports a flexible "choose-your-deliverable" model so teams can emphasize code, research/reporting, or presentation/communication depending on strengths and learning goals. Read the options below and choose one track for your team.

## 2. Deliverable Tracks (Choose one)
Pick one track and state your choice (and brief rationale) in the notebook top-level summary before Week 10.

Track A — Full Project (default)
- What to submit: complete runnable codebase, tests, README, 1000–1500 word final report, and a 10–15 minute recorded presentation or live demo.
- Expectations: full implementation of core features, automated tests, CI or build instructions, and a polished presentation.
- Points weight: same baseline rubric in Section 6 (120/120/60 split).

Track B — Report-First (research / demonstration)
- What to submit: 3000–4000 word technical report (design, evaluation, AI-involvement, security, testing strategy), a minimal runnable prototype or representative artifacts, and a 10-minute presentation (slides only or short demo).
- Expectations: deeper design & evaluation, experimental evidence, reproducible artifacts (even if partial), and strong analysis of AI usage.
- Grading emphasis: Report & analysis (50%), Technical artifacts & reproducibility (30%), Presentation (20%).

Track C — Presentation / Demo Focus
- What to submit: professional slide deck (10–15 minutes), a polished demo (recorded or live) highlighting selected features, short written summary (500–1,000 words), and link to code repository (may be prototype-level).
- Expectations: emphasis on communication, story, and usability; demo must be reproducible or clearly documented.
- Grading emphasis: Presentation & demo (60%), Technical correctness (25%), Report clarity (15%).

Notes on selection
- One team member must add a short single-paragraph justification to this notebook and commit it to the repo by Week 10.
- If a team does not declare a track by Week 10 they will be graded under Track A by default.

## 5. Capstone Tasks (What to do + Record)
For every track, the notebook must document: context, exact prompts used for design/code/docs, AI outputs, manual edits, tests run, and verification evidence (screenshots, CI logs, recorded demo). Below are core activities; adapt them to your chosen track.

1. Define project goals, scope, and technical architecture.
2. Use AI tools to assist with user stories, UML diagrams, and planning (record prompts and edits).
3. Implement core features using AI-assisted code generation; include secure-coding practices and tests where relevant.
4. Provide testing, debugging, and JavaDoc documentation with AI assistance recorded.
5. Create README with build/run instructions and a 1000–1500 word report (Track B may substitute a longer research report).
6. Prepare a 10–15 minute recorded presentation or live demo (Track C may focus more heavily on this deliverable).
7. Provide peer reviews for at least two other teams and include reflections.

Record for each item:
- Exact prompts, AI outputs, manual edits, and verification evidence (tests, screenshots, demo links).
