## Purpose of Assignment

This assignment is intended to help you:

- Analyze failure modes in your coding agent
- Evaluate reliability, security, and maintainability
- Propose improvements to increase robustness
- Strengthen the safety controls around Java tools before adding an autonomous loop


## Steps / Requirements of Assignment

Conduct a risk analysis of your current Java coding agent workflow in your existing `comp331-java-agent` repository. Focus on realistic failures in your own implementation rather than generic AI risks. Your system now has tools that can inspect files, read files, write files, and run Java evaluation commands, so your analysis should be specific to those capabilities.

1. Review each tool from Lab 3 and identify what could go wrong if an LLM requested unsafe or incorrect arguments.
2. Identify at least six failure modes. Include path traversal, accidental overwrite, command execution risk, secret exposure, misleading model output, incomplete tests, and human oversight gaps where relevant.
3. Analyze reliability, security, maintainability, and human oversight risks. At least one risk must come from each category.
4. Rate each risk by likelihood and impact using a simple scale such as Low, Medium, or High.
5. Propose concrete mitigations. Examples include path normalization, read-size limits, command allowlists, write approval checkpoints, max iteration limits, logging, and better test coverage.
6. Implement at least two small safeguards or tests based on your analysis. Reasonable options include blocking absolute paths, adding a command allowlist, adding a write confirmation mode, or adding tests for secret-like files.
7. Add a production-readiness checklist that distinguishes what is currently safe enough for coursework from what would be required before real use.
8. Identify which mitigations you plan to carry into Lab 5 and the final project.

Suggested risk table columns:

| Risk | Tool/Component | Category | Likelihood | Impact | Current Control | Proposed Mitigation |
|---|---|---|---|---|---|---|

Suggested checkpoint commands:

```bash
mvn test
mvn test -Dtest=*Tool*
```
