# Lab 4: AI Risk Analysis and Production Readiness

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Week:** 9  
**Points:** 90

This lab analyzes the system built in [Lab 3: AI System Implementation](lab03_ai_system_implementation.ipynb). Continue maintaining the [Weekly AI Interaction Log](weekly_ai_interaction_log.ipynb).


## Purpose of Assignment

This assignment is intended to help you:

- Analyze failure modes in your coding agent
- Evaluate reliability, security, and maintainability
- Propose improvements to increase robustness


## Steps / Requirements of Assignment

Conduct a risk analysis of your current coding agent workflow. Focus on realistic failures and production-readiness concerns rather than generic AI risks.

1. Identify at least five failure modes in your system.
2. Analyze reliability, security, maintainability, and human oversight risks.
3. Rate each risk by likelihood and impact.
4. Propose concrete mitigations.
5. Identify which mitigations you can implement before the final project submission.


## Action Items

- Analyze your system.
- Document findings clearly.
- Add risk analysis documentation to your repository.
- Record any AI-assisted analysis interactions in your weekly log.


## Deliverables

Your repository should include:

- Risk analysis document or README section.
- Failure mode table with likelihood, impact, and mitigation.
- Production-readiness checklist.
- Prioritized improvement plan for the final project.
