## Purpose of Assignment

This assignment is intended to help you:

- Set up your development environment for AI-assisted software engineering
- Verify **GitHub Copilot** is working correctly
- Understand the difference between **vibe coding** and structured AI-assisted development
- Introduce the concept of **programming with intent**
- Recognize key **strengths and limitations of AI-generated code**
- Interpret the **Software 2.0** perspective


## Steps / Requirements of Assignment

Complete the following setup tasks:

1. Install **JDK 17+** and verify installation with:

```bash
java -version
```

2. Install one development environment:
   - **IntelliJ IDEA**, or
   - **VS Code** with Java extensions

3. Install **Git**, sign in to **GitHub**, and confirm you can create or access a GitHub repository.

4. Install and sign in to **GitHub Copilot**.

5. Create a simple Java project named:

```text
ai-swe-course
```

6. Add a basic `Main.java` file.

7. Verify Copilot suggestions are working:
   - Type a simple comment such as:

```java
// method to add two numbers
```

   - Confirm Copilot ghost text appears.
   - Accept at least one suggestion.

8. Generate **one simple method** using Copilot.

Examples:
- Add two numbers
- Reverse a string
- Check if a number is even

Keep the code simple. The goal is to confirm your setup works.
