## 5. Prompting Exercises (Experiments)
For each experiment, **paste exact prompts** and record observations in the **Prompt Journal** (Section 8). Keep code minimal—stubs and short methods are sufficient.

### Experiment 1 — Comment-to-Code Basics (Inline)
**Goal:** Use short comments to elicit simple Java methods and compare suggestions.

Pick **two** small targets (choose any):
- `boolean isAnagram(String a, String b)` (ignore case/whitespace)
- `int clamp(int value, int min, int max)`
- `String normalizeEmail(String s)` (trim, lowercase, basic structure)
- `int[] dedupeSorted(int[] a)` (in-place or new array—state preference in comment)

**What to do:**
- Write a **single-line** comment above a stub describing the behavior.
- Trigger Copilot and **cycle through alternatives**.
- Accept one, then try a **multi-line** comment adding 2–3 edge cases. Compare.
- Note what changed and why.

**Record:** prompt text(s), which suggestion you accepted and why, edge cases mentioned, and a short self-check (does it match intent?).

### Experiment 2 — Docstring-Driven Style & Constraints
**Goal:** Steer Copilot on naming, Javadoc, immutability, and complexity hints using **docstrings/Javadoc**.

Pick **one** target:
- `List<Integer> slidingWindowMax(int[] nums, int k)` (hint at O(n) desired)
- `String toTitleCase(String s, Locale locale)` (rules for apostrophes/hyphens)

**What to do:**
- Write a Javadoc block specifying **pre/postconditions**, **constraints** (e.g., O(n) time, no mutable statics), and **three examples**.
- Generate code with Copilot.
- If style/constraints are ignored, **tighten the Javadoc** and regenerate.

**Record:** evolution of the Javadoc prompt, which constraints were ignored/obeyed, and a brief readability assessment.

### Experiment 3 — Explanations & Critiques (Chat or Inline)
**Goal:** Ask Copilot to **explain** its code and then to **critique & improve** it (naming, edge cases, readability).

**What to do:**
- Ask: “Explain what this method does; call out assumptions and potential edge cases.”
- Follow-up: “Propose a cleaner version that improves <X> without changing behavior; show a diff or new method.”
- Optionally request a **checklist comment** at the top indicating invariants Copilot claims to uphold.

**Record:** original vs. improved versions (summary), issues found, and whether the critique actually improved clarity or safety.

## Appendix B — Starter Prompt Patterns & Anti-patterns
**Pattern: Comment-to-Code (Inline)**
```java
// Return true if a and b are anagrams (ignore case/whitespace). Include 3 edge cases in code comments.
```

**Pattern: Javadoc Spec (Constraints & Examples)**
```java
/**
 * Returns the max in each sliding window of size k.
 * Constraints: O(n) time; no mutable statics; handle k==1 and k==n.
 * Examples:
 * [1,3,-1,-3,5,3,6,7], k=3 -> [3,3,5,5,6,7]
 * [2], k=1 -> [2]
 * [ ], k invalid -> define behavior (throw or return empty) and implement.
 */
```

**Pattern: Critique & Improve**
```
Explain what this method does, list edge cases it misses, and suggest a clearer version without changing behavior.
```

**Anti-patterns to avoid**
- Vague asks ("make it better") without criteria.
- Omitting edge cases and then trusting defaults.
- Overlong prompts that bury key constraints; prefer bullet-style clarity.
