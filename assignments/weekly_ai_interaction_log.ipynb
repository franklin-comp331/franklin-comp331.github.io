# Weekly AI Interaction Log

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Weeks:** 2, 3, 4, 5, 6, 7, 8, 9, 10  
**Points:** 10 points each week / 90 total


## Purpose of Assignment

This assignment is intended to help you learn to do the following:

- Develop intentional and structured use of GitHub Copilot or another approved AI/LLM tool
- Reflect on how prompting, iteration, and evaluation affect outcomes
- Identify patterns in successful and unsuccessful AI interactions
- Build awareness of your own engineering decision-making process
- Document how AI-assisted development evolves over time


## Steps / Requirements of Assignment

Submit a separate AI interaction reflection for Weeks 2 through 10. This is the same recurring assignment, but every submission should cover only the current course week. Focus on the AI-assisted work, lab activity, reading, debugging, testing, or project development completed during the week.

Create a new document for every submission. Do not submit one cumulative document containing multiple weeks unless your professor explicitly asks for it. Submit the reflection through Canvas as a document or PDF, such as `.docx` or `.pdf`, unless your professor approves another format.

Your entry should include:

1. Week number and date
2. Development task or learning activity
3. Your intent before using Copilot
4. Copilot mode used: inline suggestion, chat, edit, agent/workspace, test generation, review, or debugging
5. Prompt, comment, or interaction summary
6. AI tool output or suggestion summary
7. What you accepted, modified, or rejected
8. Verification method: test, compiler, manual review, comparison, or rejected output
9. Reflection on what worked, what failed, and what you will change next time

Suggested template:

```text
Intent:
Copilot mode used:
Context I gave Copilot:
Copilot output:
What I accepted:
What I changed:
How I verified it:
What I learned:
```
