# Weekly AI Interaction Log

**Course Name/Number:** COMP 331: AI-Assisted Software Engineering  
**Weeks:** 2, 3, 4, 5, 6, 7, 8, 9, 10  
**Points:** 10 points each week / 90 total


## Purpose of Assignment

This assignment is intended to help you learn to do the following:

- Develop intentional and structured use of GitHub Copilot
- Reflect on how prompting, iteration, and evaluation affect outcomes
- Identify patterns in successful and unsuccessful AI interactions
- Build awareness of your own engineering decision-making process
- Document how AI-assisted development evolves over time


## Steps / Requirements of Assignment

Maintain a running AI interaction log from Week 2 through Week 10. A single document, a folder of weekly Markdown files, or a dedicated section in your course repository is acceptable.

Each weekly entry should include:

1. Week number and date
2. Development task or learning activity
3. Your intent before using Copilot
4. Prompt, comment, or interaction summary
5. Copilot output or suggestion summary
6. What you accepted, modified, or rejected
7. How you evaluated the result
8. Reflection on what worked, what failed, and what you will change next time


## Action Items

- Create a running AI interaction log in your repository.
- Add one entry each week from Week 2 through Week 10.
- Record enough detail that another reader can understand your intent, the AI interaction, and your engineering decision.
- Submit the weekly entry according to your professor's Canvas instructions.


## Deliverables

Submit a link to the current week's log entry. Recommended repository structure:


## Grading

Each weekly entry is worth 10 points.

| Category | Points |
|---|---:|

| **Weekly Total** | **10** |

Across Weeks 2-10, the assignment is worth 90 points total.
