## 1. Overview
Investigate ethical, legal, and societal implications of AI-assisted software engineering. Through scenario analysis, a case study, and a reflective essay, you will identify risks (bias, IP, accountability) and propose mitigation strategies.

---

## 2. Alignment with the Syllabus
- Analyze ethical challenges in AI-assisted development.
- Propose mitigation and governance strategies.
- Communicate findings in a clear, evidence-backed report.

---

## 3. Tooling & Policy
- Use GitHub Copilot for brainstorming or drafting but record prompts and attribute assistance.
- Ethical analyses must cite sources; do not submit proprietary or confidential data.

---

## 4. Tasks (What to do + Record)

### Task 1 — Ethical Scenarios Review (20 pts)
What to do:
- Review 2–3 AI-generated code samples or incidents illustrating bias, IP, or accountability problems.
Record:
- Scenario descriptions, prompts (if applicable), and evidence supporting your analysis.

### Task 2 — Analysis & Mitigation (25 pts)
What to do:
- For each scenario propose concrete mitigation strategies (technical, process, policy).
Record:
- Actionable mitigation steps and evaluation of their trade-offs.

### Task 3 — Reflection Essay (25 pts)
What to do:
- Write a 500-word essay on balancing innovation and responsibility in AI-assisted development.
Record:
- Final essay text and key references.

### Task 4 — Case Study Research (25 pts)
What to do:
- Select a real-world case related to AI-assisted development; analyze tools, workflow, outcomes, and lessons.
Record:
- Case summary, sources, and your assessment of effectiveness and risk.

### Task 5 — Future Trends Exploration (20 pts)
What to do:
- Explore emerging technologies and predict their impact on software engineering.
Record:
- Short write-up with references and potential opportunities/risks.

### Task 6 — Group Presentation (20 pts)
What to do:
- Prepare a collaborative presentation summarizing findings.
Record:
- Link to slide deck or video and brief notes on contributions.
