## 5. Tasks (What to do + Record)

### Task 1 — Code Security Review (25 pts)
What to do:
- Select AI-generated Java code and analyze for vulnerabilities (injection, insecure error handling, unsafe deserialization, etc.).
Record:
- Exact prompts used, identified issues with rationale, and evidence (code snippets).

### Task 2 — Secure Refactoring (25 pts)
What to do:
- Refactor code to address identified vulnerabilities; use Copilot suggestions but validate manually.
Record:
- Before/after code, tests that demonstrate the fix, and rationale for chosen fixes.

### Task 3 — Safe Coding Documentation (20 pts)
What to do:
- Summarize secure-coding principles applied and document remaining risks.
Record:
- Short checklist or README section explaining the secure choices made.

### Task 4 — AI-Assisted Project Planning (30 pts)
What to do:
- Use Copilot to propose a small project plan: deliverables, milestones, and 3–5 user stories.
Record:
- Prompts and final plan; comment which parts you edited and why.

### Task 5 — AI in Task Management (15 pts)
What to do:
- Demonstrate one example of using AI to generate or organize sprint tasks.
Record:
- Prompt(s) and produced tasks; a short evaluation of usefulness and risks.

### Task 6 — Reflection (20 pts)
What to do:
- Reflect on AI’s role in security and project planning (≈300 words).
Record:
- Reflection text focusing on trade-offs and recommended human oversight patterns.
